### Polynomial Regression

```python
""" Importing libraries """
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# importing dataset
dataset = pd.read_csv("Position_Salaries.csv")
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values

dataset
```

```python
""" Traning the Linear Regression Model on the whole dataset """

from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
""" Traning the Polynomial Regression model on the whole dataset """

from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

```python
""" Visualising the Linear Regression results """
plt.scatter(X, y, color="blue")
plt.plot(X, lin_reg.predict(X), color="red")
plt.title("Truth or bluff (Linear Regression)")
plt.xlabel("Position level")
plt.ylabel("Salary")
plt.show()
```

```python
""" Visualising the Polynomial Regression results """
plt.scatter(X, y, color="blue")
plt.plot(X, lin_reg_2.predict(X_poly), color="red")
plt.title("Truth or bluff (Polynomial Regression)")
plt.xlabel("Position level")
plt.ylabel("Salary")
plt.show()
```

```python
""" Visualising the Polynomial Regression Result (for higher resolution and smoother curve) """
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color="blue")
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color="red")
plt.title("Truth or bluff (Polynomial Regression)")
plt.xlabel("Position level")
plt.ylabel("Salary")
plt.show()
```

```python
""" Predicting a new result with Linear Regression """
lin_reg.predict([[6.5]])
```

```python
""" Predicting a new result with Polynomial Regression """
test_arr = poly_reg.fit_transform([[6.5]])
lin_reg_2.predict(test_arr)
```
