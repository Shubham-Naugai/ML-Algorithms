```python
""" Simple Linear Regression """

# importing libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# importing dataset
dataset = pd.read_csv("Salary_Data.csv")
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
""" Splitting the dataset into the Traning set and Test set """
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
""" Predicting the test set """
y_pred = regressor.predict(X_test)
```

```python
""" Visualising the training set """
plt.scatter(X_train, y_train, color="red")
plt.plot(X_train, regressor.predict(X_train), color="blue")
plt.title("Salary vs Experience (Traning Set)")
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
""" Visualising the test set """
plt.scatter(X_test, y_test, color="red")
plt.plot(X_train, regressor.predict(X_train), color="blue")
# plt.plot(X_test, regressor.predict(X_test), color="blue")  # regression will same as traning set
plt.title("Salary vs Experience (Test Set)")
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
""" Using Gradient descent """

""" Batch Gradient Descent """

X = X_train     # Independent variable
y = y_train     # dependent variable

plt.plot(X, y, 'bo')
```

```python
# y = mX + c

m = 0
c = 0
n = len(y)

lr = 0.02
epochs = 1000

def predY(m, c):
  y_pred = m*X + c
  return y_pred

for i in range(epochs): 
    y_pred = predY(m,c)
    D_m = 0
    D_c = 0
    for i in range(n):
      D_m += X[i] * (y[i] - y_pred[i])
      D_c += y[i]- y_pred[i]
    Dm = -2*(D_m/n)
    Dc = -2*(D_c/n)
    m = m - (lr * Dm)                       # Update m
    c = c - (lr * Dc)                       # Update c
print(m,c)
```

```python
# Making predictions
y_pred = m*X + c

plt.scatter(X, y)
plt.plot(X, y_pred, color='red')  # regression line
plt.show()
```
